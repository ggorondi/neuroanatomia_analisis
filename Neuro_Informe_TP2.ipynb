```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#armar dataframes con las 3 tablas
contraste = pd.read_csv('/content/contraste.csv',delimiter=";") 
perspectiva = pd.read_csv('/content/perspectiva.csv',delimiter=";") 
rpc = pd.read_csv('/content/relleno_punto_ciego.csv',delimiter=";") 
```

```python
#Pregunta 1
respuestas_color=['respuestas_rojo', 'respuestas_verde', 'respuestas_ns']
sinpriming1=[rpc[rpc["priming_color"]=="nada"][x].values.sum() for x in respuestas_color]
plt.bar(respuestas_color, sinpriming1)
plt.title('Preferencia de color sin priming')
plt.ylabel("Frecuencia absoluta")

```

```python
#pregunta 2
respuestas_direcciones=["horizontal","vertical","respuestas_ns"]
sinpriming2=rpc[rpc["priming_color"]=="nada"]
respuestas=[sinpriming2[sinpriming2["horizontal"]=="verde"]["respuestas_verde"].values.sum()+sinpriming2[sinpriming2["horizontal"]=="rojo"]["respuestas_rojo"].values.sum(),sinpriming2[sinpriming2["vertical"]=="verde"]["respuestas_verde"].values.sum()+sinpriming2[sinpriming2["vertical"]=="rojo"]["respuestas_rojo"].values.sum(),sinpriming2["respuestas_ns"].values.sum()]
plt.bar(respuestas_direcciones, respuestas)
plt.title('Preferencia de direccionalidad sin priming')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 3 a
conprimingrojo=[rpc[rpc["priming_color"]=="rojo"][x].values.sum() for x in respuestas_color]
plt.bar(respuestas_color, conprimingrojo)
plt.title('Preferencia de color con priming rojo')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 3 b
conprimingverde=[rpc[rpc["priming_color"]=="verde"][x].values.sum() for x in respuestas_color]
plt.bar(respuestas_color, conprimingverde)
plt.title('Preferencia de color con priming verde')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 3 c
plt.bar(["color primeado","color no primeado","no se"], [conprimingrojo[0]+conprimingverde[1],conprimingrojo[1]+conprimingverde[0],conprimingrojo[2]+conprimingverde[2]])
plt.title('Preferencia de color con priming')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 4 a
conprimingvertical=rpc[rpc["priming_color.1"]=="vertical"]
respuestasprimingverti=[conprimingvertical[conprimingvertical["vertical"]=="verde"]["respuestas_rojo"].values.sum()+conprimingvertical[conprimingvertical["vertical"]=="rojo"]["respuestas_verde"].values.sum(),conprimingvertical[conprimingvertical["vertical"]=="verde"]["respuestas_verde"].values.sum()+conprimingvertical[conprimingvertical["vertical"]=="rojo"]["respuestas_rojo"].values.sum(),conprimingvertical["respuestas_ns"].values.sum()]
plt.bar(["horizontal","vertical","no se"], respuestasprimingverti)
plt.title('Preferencia de direccin con priming vertical')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 4 b
conpriminghorizontal=rpc[rpc["priming_color.1"]=="horizontal"]
respuestaspriminghorizontal=[conpriminghorizontal[conpriminghorizontal["horizontal"]=="verde"]["respuestas_verde"].values.sum()+conpriminghorizontal[conpriminghorizontal["horizontal"]=="rojo"]["respuestas_rojo"].values.sum(),conpriminghorizontal[conpriminghorizontal["horizontal"]=="verde"]["respuestas_rojo"].values.sum()+conpriminghorizontal[conpriminghorizontal["horizontal"]=="rojo"]["respuestas_verde"].values.sum(),conpriminghorizontal["respuestas_ns"].values.sum()]
plt.bar(["horizontal","vertical","no se"], respuestaspriminghorizontal)
plt.title('Preferencia de direccin con priming horizontal')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 4 c
plt.bar(["direccin primeada","direccin no primeada","no se"], [respuestasprimingverti[1]+respuestaspriminghorizontal[0],respuestasprimingverti[0]+respuestaspriminghorizontal[1],respuestasprimingverti[2]+respuestaspriminghorizontal[2]])
plt.title('Preferencia de direccin con priming')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 7
respuestas_palos=["Convexa","Concava","Iguales"]
iguales=perspectiva[perspectiva["mayor"]=="iguales"]
respuestas=[iguales[iguales["convexo"]=="B"]["respuesta_B"].values.sum()+iguales[iguales["convexo"]=="A"]["respuesta_A"].values.sum(),iguales[iguales["convexo"]=="B"]["respuesta_A"].values.sum()+iguales[iguales["convexo"]=="A"]["respuesta_B"].values.sum(),iguales["respuesta_ns"].values.sum()]
plt.bar(respuestas_palos, respuestas)
plt.title('Eleccin de barra mayor con barras iguales')
plt.ylabel("Frecuencia absoluta")
```

```python
#pregunta 8
respuestas_palos=["Convexa","Concava","Iguales"]
iguales=perspectiva[perspectiva["mayor"]!="iguales"]
iguales
respuestas=[iguales[iguales["convexo"]=="B"]["respuesta_B"].values.sum()+iguales[iguales["convexo"]=="A"]["respuesta_A"].values.sum(),iguales[iguales["convexo"]=="B"]["respuesta_A"].values.sum()+iguales[iguales["convexo"]=="A"]["respuesta_B"].values.sum(),iguales["respuesta_ns"].values.sum()]
plt.bar(respuestas_palos, respuestas)
plt.title('Eleccin de barra mayor con barra convexa mayor')
plt.ylabel("Frecuencia absoluta")
```

```python
#extra
contrario=0
ese=0
for index in range(len(perspectiva["mayor"].values)-1):
  if perspectiva.loc[index,"mayor"]!="iguales" and perspectiva.loc[index+1,"mayor"]=="iguales":
    ese+=perspectiva.loc[index+1,f"respuesta_{perspectiva.loc[index,'mayor']}"]
    if perspectiva.loc[index,"mayor"]=="A":
      contrario+=perspectiva.loc[index+1,f"respuesta_{'B'}"]
    else:
      contrario+=perspectiva.loc[index+1,f"respuesta_{'A'}"]
print(contrario,ese)

plt.bar(["El mismo", "El contrario"], [ese, contrario])
plt.title('Eleccin de barra en base a la vez previa')
plt.ylabel("Frecuencia absoluta")
```

```python
#extra2
respuestas_palos=["Convexa","Concava","Iguales"]
iguales=perspectiva[perspectiva["mayor"]=="iguales"]
iguales_primermitad=iguales.head(4)
iguales_segundamitad=iguales.tail(4)
respuestas=[iguales_primermitad[iguales_primermitad["convexo"]=="B"]["respuesta_B"].values.sum()+iguales_primermitad[iguales_primermitad["convexo"]=="A"]["respuesta_A"].values.sum(),iguales_primermitad[iguales_primermitad["convexo"]=="B"]["respuesta_A"].values.sum()+iguales_primermitad[iguales_primermitad["convexo"]=="A"]["respuesta_B"].values.sum(),iguales_primermitad["respuesta_ns"].values.sum()]
respuestas2=[iguales_segundamitad[iguales_segundamitad["convexo"]=="B"]["respuesta_B"].values.sum()+iguales_segundamitad[iguales_segundamitad["convexo"]=="A"]["respuesta_A"].values.sum(),iguales_segundamitad[iguales_segundamitad["convexo"]=="B"]["respuesta_A"].values.sum()+iguales_segundamitad[iguales_segundamitad["convexo"]=="A"]["respuesta_B"].values.sum(),iguales_segundamitad["respuesta_ns"].values.sum()]
X_axis = np.arange(len(respuestas))
fig, ax = plt.subplots(1,1)
plt.bar(X_axis - 0.2, respuestas, 0.4, label = "Primer mitad del experimento")
plt.bar(X_axis + 0.2, respuestas2, 0.4, label = 'Segunda mitad del experimento')
plt.legend()
plt.title("Primer mitad vs segunda mitad del experimento")
ax.set_xticks(X_axis)
plt.ylabel("Frecuencia absoluta")
ax.set_xticklabels(respuestas_palos, fontsize=10)
```
